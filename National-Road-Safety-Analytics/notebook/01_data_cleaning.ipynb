```python
import pandas as pd
```

# Importing the all the datasets

```python
micro = pd.read_csv("data/raw/accident_prediction_india.csv")
ra19 = pd.read_csv("data/raw/RA2019_A2.csv")
ra20 = pd.read_csv("data/raw/RA2020_A1.csv")
summary = pd.read_csv("data/raw/national_summary.csv")
```

Cleaning Micro Accident Dataset

```python
micro.columns = micro.columns.str.strip().str.lower().str.replace(" ", "_")

micro = micro.drop_duplicates()
micro = micro.dropna(subset=["state_name", "number_of_casualties"])
micro["number_of_casualties"] = micro["number_of_casualties"].astype(int)

micro.to_csv("data/cleaned/clean_micro_accidents.csv", index=False)
```

Cleaning Govt Summary Data

```python
ra19.columns = ra19.columns.str.lower().str.replace(" ", "_")
ra20.columns = ra20.columns.str.lower().str.replace(" ", "_")

ra19 = ra19.fillna(0)
ra20 = ra20.fillna(0)

ra19.to_csv("data/cleaned/clean_ra2019.csv", index=False)
ra20.to_csv("data/cleaned/clean_ra2020.csv", index=False)
```

```python
summary.columns = summary.columns.str.lower().str.replace(" ", "_")
summary = summary.fillna(0)

summary.to_csv("data/cleaned/clean_national_summary.csv", index=False)
```

Phase-1 involved ingestion and cleaning of official MoRTH and Government of India road accident datasets.
Data normalization, duplicate removal, missing value handling and schema standardization were performed using Python (pandas).
